```python
import prolif as plf
import MDAnalysis as mda
```

```python
protein = mda.Universe("../proteinpreparation/protein_prepared/prepared_5vfi.pdb")

#!obabel -isdf btk_fenebrutinib_analogs.sdf_i0-again2_out1.sdf -O btk_fenebrutinib_analogs.sdf_i0-again2_out1.pdb

```

```python
from openbabel import pybel

# Use pybel to read the SDF, add hydrogens, and save as PDB
mol = next(pybel.readfile("sdf", "btk_fenebrutinib_analogs.sdf_i0-again1_out1_vina_out.sdf"))
mol.addh()  # Add hydrogens
mol.write("pdb", "btk_fenebrutinib_analogs.sdf_i0-again2_out1.pdb", overwrite=True)
ligand = mda.Universe("btk_fenebrutinib_analogs.sdf_i0-again2_out1.pdb")
```

```python
protein_plf = plf.Molecule.from_mda(protein)
poses_plf = plf.sdf_supplier("btk_fenebrutinib_analogs.sdf_i0-again1_out1_vina_out.sdf")
```

```python
fp = plf.Fingerprint(count=True)
```

```python
# run on your poses
fp.run_from_iterable(poses_plf, protein_plf)
```
